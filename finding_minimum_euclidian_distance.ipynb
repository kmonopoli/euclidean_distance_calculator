### The purpose of the script is to calculate the smallest distance between an object in dataset 1 and an object in dataset 2

```python
import pandas as pd
# read in the data
# dataset 1
d1 = pd.read_csv('Statistics for 076 01102022_stack.tif (green)_dataset1.csv')
display(d1)

# dataset 2
d2 = pd.read_csv('Statistics for 076 01102022_stack.tif (red) dataset2.csv')
display(d2)

```

```python
import numpy as np
def compute_3d_dist(p1, p2):
    '''Compute the Euclidean (as the crow flies) Distance between two coordinates in three-dimmensional space
    Keyword arguments:
    INPUT:
        p1 -- numpy array of the three-point coordinate p1 
            (constructed by: np.array([x1, y1, z1]) where x1, y1, and z1 are floats)

        p2 -- numpy array of the three-point coordinate p2 
            (constructed by: np.array([x2, y2, z2]) where x2, y2, and z2 are floats) 

    OUTPUT:
        dist -- a float represeting the Euclidean Distance between points p1 and p2
    
    '''
    squared_dist = np.sum((p1-p2)**2, axis=0)
    dist = np.sqrt(squared_dist)
    return dist


# # TESTING: compute_3d_dist (uncomment asserts to test - also included prints of the same things so you can test with those as well)
# # Data input representations:
# # a) as inidividual points
# x_1 = 3.0
# y_1 = 4.0
# z_1 = 5.0

# x_2 = 9.0
# y_2 = 7.0
# z_2 = 3.0

# # b) as lists of floats
# pt_1_ = [x_1,y_1,z_1]
# pt_2_ = [x_2,y_2,z_2]

# # c) as numpy arrays 
# pt_1 = np.array([x_1,y_1,z_1])
# pt_2 = np.array([x_2,y_2,z_2])

# # Testing various input types (should all give the same output)
# # 1) floats directly 
# print(compute_3d_dist(np.array([3.0,4.0,5.0]), np.array([9.0,7.0,3.0])))
# # 2) coordinates as a list
# print(compute_3d_dist(np.array([x_1,y_1,z_1]), np.array([x_2,y_2,z_2])))
# # 3) numpy arrays 
# print(compute_3d_dist(pt_1,pt_2))
# assert ((compute_3d_dist(np.array([3.0,4.0,5.0]), np.array([9.0,7.0,3.0]))), (compute_3d_dist(np.array([x_1,y_1,z_1]), np.array([x_2,y_2,z_2]))), (compute_3d_dist(pt_1,pt_2))) == (7.0, 7.0, 7.0), "Should all be 7.0"

# # Testing computation veracity 
# # 4) Compute distance between the same point (should be zero)
# print(compute_3d_dist(pt_1,pt_1))
# assert compute_3d_dist(pt_1,pt_1) == 0, "Should be 0"

# # 5) Compute distances between 2 points and compare to calculator:
# #                https://www.calculatorsoup.com/calculators/geometry-solids/distance-two-points.php
# #     For this example ([7, 4, 3] to [17, 6, 2]) should be  10.246951
# pt_1_ = [7, 4, 3]
# pt_2_ = [17, 6, 2]
# print(compute_3d_dist(np.array(pt_1_),np.array(pt_2_))) 
# assert compute_3d_dist(np.array(pt_1_),np.array(pt_2_)) == 10.246950765959598, "Should be 10.246950765959598"


```

```python
coord_names = ['X','Y','Z'] # **This might be wrong**, name of columns for your coordinates in each file. This also assumes each file has the same column names
# walk row by row through dataset 1 reading in coordinates and then comparing with each row of dataset 2
# loops through each row of dataset 1


# holds the minimum distances in a list of floats
min_dist_ls = []
# holds the dataset 2 coordinates that provide the minimum distances with each dataset 1 coordinate in a list (with each element in the list being a numpy array to represent the single coordinate)
d2_closest_coord_ls = []

# for i in [0,1,2,3]: # NOTE: you can uncomment this line and comment out the line below to test with a smaller dataset
for i in range(len(d1)): # NOTE: this type of for loop does not require incrementing the counter (j) so if you call j+=1 anywhere in your code it could cause some rows to be skipped
    
    # extracts the coordiantes for row at index i in dataset 1
    # coordinates are extracted in the order they appear in coord_names (so X,Y,Z)
    # converts these coordinates into a numpy array so it is appropriate input for compute_3d_dist function
    d1_coord = np.array(d1.iloc[i][coord_names])
   
    # holds the minimum distance to compare as we walk through the points
    # initially set to some very large float (number with a decimal) here set to infinity
    min_dist = float('inf')
    
    # holds the coordinate for the minimum distance as a np.array (to start it is just an empty list)
    d2_closest_coord = np.array([])
    
    # loops through each row of dataset 2 
    #       NOTE: this is a nested loop so it will run i*j times (thus potentially a LONG time)
    #for j in [0,1,2,3]: # NOTE: you can uncomment this line and comment out the line below to test with a smaller dataset
    for j in range(len(d2)): # NOTE: this type of for loop does not require incrementing the counter (j) so if you call j+=1 anywhere in your code it could cause some rows to be skipped 
        
        # extracts the coordiantes for row at index j in dataset 2
        # coordinates are extracted in the order they appear in coord_names (so X,Y,Z)
        # converts these coordinates into a numpy array so it is appropriate input for compute_3d_dist function
        d2_coord = np.array(d2.iloc[j][coord_names])
        
        # compute the distance between d1_coord and d2_coord
        d1_d2_dist = compute_3d_dist(d1_coord, d2_coord)
        
        # if distance between d1_coord and d2_coord is smaller than the current min_dist:
        # store the new value as min_dist and also store the d2 coordinates in d2_closest_coord
        if d1_d2_dist < min_dist:
            min_dist = d1_d2_dist
            d2_closest_coord = d2_coord
    
    
    # store the minimum distance and coordinate
    min_dist_ls.append(min_dist)
    d2_closest_coord_ls.append(d2_closest_coord)
        
        



# Output the data:

# 1) coordinates of every point in dataset 1 each as a single column
out_data_1 = d1[coord_names]
out_data_1.columns = ['X_dataset_1','Y_dataset_1','Z_dataset_1']

# 2) coordinates of the corresponding closest point in dataset 2 each as a single column
out_data_2 = pd.DataFrame(d2_closest_coord_ls)
out_data_2.columns = ['X_closest_pt_dataset_2','Y_closest_pt_dataset_2','Z_closest_pt_dataset_2']

# 3) combine data
out_data = pd.concat([out_data_1, out_data_2],axis=1)

# 4) distances between the points
out_data['distance'] = min_dist_ls

# 5) save data to csv file
fname = 'min_distances_dataset-1_to_dataset-2.csv'
out_data.to_csv(fname)
print("data saved to:\n",fname)

display(out_data)     
        

```
